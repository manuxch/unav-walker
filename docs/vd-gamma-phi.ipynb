```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import root_scalar

g = 1
omega = 6.95276344
mu = 0.17

def Theta(Gamma):
    return 2 * np.arcsin(np.pi * mu / Gamma)

def vd(Gamma, phi):
    return g * Gamma / (4 * omega) * np.cos(phi - Theta(Gamma))
```

```python
phi_data = np.linspace(0, 6.3, 100)
plt.plot(phi_data, vd(1, phi_data), label=r'$\Gamma = 1$')
plt.plot(phi_data, vd(3, phi_data), label=r'$\Gamma = 3$')
plt.legend()
plt.xlabel(r'$\phi$')
plt.ylabel(r'$v_d$')
plt.show()
# Proporcionar un valor inicial
#x0 = 0.5

# Resolver la ecuaci贸n#
#sol = fsolve(f, x0

#print(f"Soluci贸n encontrada: x = {sol[0]}")

```

```python
def vd_sol(phi):
    return vd(3, phi) - vd(1, phi)

sol = root_scalar(vd_sol, bracket=[1, 2], method='bisect')
print(f"Soluci贸n: phi = {sol.root}")
```
